# Case Study 3: Pandas, Seaborn
## Problem Statement:
The dataset contains Happiness Score for 153 countries along with the factors used to explain the score. Create visualizations for the following using Seaborn.
- Scatter plot to show the relationship between GDP per capita and happiness score.
- Heat map to show the correlations between different variables.
- Bar plot to show the Happiness score for the top 10 countries.
- Box plot to show the distribution of each variable in the dataset.
- Pair Plot to show the relationship between each variable in the dataset.

Dataset: happiness_report.csv


```python
df = pd.read_csv("happiness_report.csv")
df.head(15)
```

## Q1. Scatter plot to show the relationship between GDP per capita and happiness score.

```python
plt.figure(dpi=100)
sns.scatterplot(x="GDP per capita", y="Score", data=df, hue='Healthy life expectancy', palette='mako_r')
plt.title("Happiness Score v/s GDP per Capita")
plt.show()
```

### We can observe that as most countries with High Happiness Scores have High GDP per Capita 

## Q2. Heat map to show the correlations between different variables.


```python
plt.figure(dpi=100)
sns.heatmap(df.corr(), annot=True, cmap='mako')
plt.title("Relationship between Variables ")
plt.show()
```

### We can observe that features like overall rank, score, gdp, healthy life expectancy are related to each other.


## Q3. Bar plot to show the Happiness score for the top 10 countries.


```python
plt.figure(dpi=100)
sns.barplot(y="Country or region",
           x="Score",
           data=df.head(10),
           palette="mako_r")
plt.title("Top 10 Countries with Highest Happiness Score")
plt.show();
```

### We can observe that Finland has the highest Happiness Score
